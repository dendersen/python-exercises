```python
def LIX(text: str):
  words = wordList(text)
  O = len(words)
  print("words: " + str(O))
  P = dots(text)
  print("sentences: " + str(P))
  L = long(words, 6)
  print("long words: " + str(L))
  return (O/P) + ((L*100)/O)

def wordList(text: str):
  text = text.split(" ")
  words = []
  for word in text:
    if not word == "":
      words.append(word)
  return (words)

def dots(text: str):
  return (len(text.split("."))-1)

def long(words: list, length: int):
  counter = 0
  for word in words:
    if len(word) > length:
      counter += 1
  return counter

def loadFile(fileName: str, fileExtention: str):
  with open(fileName + "." + fileExtention) as f:
    lines = f.read()
  return lines

def LIX_Translate(LIX):
  if LIX >= 55:
    return "Meget svær, faglitteratur på akademisk niveau, lovtekster."
  elif LIX > 45:
    return "Svær, f.eks. saglige bøger, populærvidenskabelige værker, akademiske udgivelser."
  elif LIX > 35:
    return "Middel, f.eks. dagblade og tidsskrifter."
  elif LIX > 25:
    return "Let for øvede læsere, f.eks. ugebladslitteratur og skønlitteratur for voksne."
  else: #LIX < 25
    return "Let tekst for alle læsere, f.eks. børnelitteratur."

def lixFromFile(fileName: str,fileExtencion: str):
  lix = LIX(loadFile(fileName,fileExtencion))
  print (lix)
  print (LIX_Translate(lix))

print("gym opgave:")
lixFromFile("load","txt")

print("\nfolkeskole opgave:")
lixFromFile("load2","txt")

# print("\n\ncontroll\n" + "words: 25\n" + "sentences: 2\n" + "long words: 5\n" + "answer: 32.5\n")
# print (LIX("wgwefyvu we wr. wtd helloweggf rqwio gfwqtfewu, qfdgugu etipuggb wqej we gi wvyi diq ein dvni feok ewo qo feno ewji feip jifeow vrni fei."))
```

```python
import numpy as np

def tabel(startingNumber,lengthOfTable):
  a = []
  for i in range(1,lengthOfTable+1):
    a.append(startingNumber * i)
  return a

def bigTabel(startingNumber,endingNumber):
  out = ""
  length = len(str(endingNumber * endingNumber)) + 1
  for i in range(startingNumber,endingNumber+1):
    a = tabel(i,endingNumber)
    for j in a:
      j = str(j)
      while len(str(j)) < length:
        j += " "
      out += j
    out += "\n"
  return(out)

print(bigTabel(1,10))
```
